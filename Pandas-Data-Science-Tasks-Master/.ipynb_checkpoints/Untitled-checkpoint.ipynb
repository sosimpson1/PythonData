```python
import pandas as pd
import os
```

```python
# Merge the 12 Months of sales data into a single file
```

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_Data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.csv("all_data.csv", index=False)
```

```python
all_data = pd.read_csv("./Output/all_data.csv")
all_data.head()
```

```python
#### Augment data with additional column
```

```python
#### ADD month column#
```

```python
#### What was the best month for sales and how much money was made
```

```python
#### Month Column
```

```python
all_data['Month'] = all_data['Order Date']
all_data.head()
```

```python
#### Clean up data get of all the NAN rows
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data.dropna()
```

```python
### Find "Or" and fix it
```

```python
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
```

```python
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
####Add a sales column
```

```python
#### Convert column tp the correct type
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) # Make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) # Make float
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
#### Best Month ...How Much was earned at the end of the mont
```

```python
def get_city(address):
    return address.split(',')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data.head()
```

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
